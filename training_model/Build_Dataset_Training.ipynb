# Read File CSV Dataset Training

```python
import pandas as pd

dataset_tweet_emotion   = pd.read_csv('dataset/mentah/Dataset Label Emotion.csv')
dataset_keyword_banjir  = pd.read_csv('dataset/mentah/Dataset Label Keyword Banjir Sumatra.csv')
dataset_keyword_jokowi  = pd.read_csv('dataset/mentah/Dataset Label Keyword Jokowi.csv')
dataset_pilpres         = pd.read_csv('dataset/mentah/Dataset Label Pilpres 2024.csv')
dataset_ppkm            = pd.read_csv('dataset/mentah/Dataset Label PPKM.csv', sep=None, engine='python')
```

# Integration Data File Emotion Tweet to Sentiment Tweet

```python
dataset_tweet_emotion
```

```python
# rename column

dataset_tweet_emotion = dataset_tweet_emotion[['tweet','label']].rename(columns={'tweet':'Tweet'})
dataset_tweet_emotion
```

```python
# mapping emotion to sentiment

emotion_to_sentiment = {
    'happy': 'positive',
    'love': 'positive',
    'anger': 'negative',
    'sadness': 'negative',
    'fear': 'negative'
}

# change label emotion to sentiment
dataset_tweet_emotion['label'] = dataset_tweet_emotion['label'].map(emotion_to_sentiment)
dataset_tweet_emotion
```

# Integration Data File Tweet Banjir Sumatra

```python
dataset_keyword_banjir
```

```python
# rename column chat and rate

dataset_keyword_banjir = dataset_keyword_banjir[['chat','rate']].rename(columns={'chat':'Tweet',
                                                                             'rate' : 'label'})
dataset_keyword_banjir
```

```python
# mapping lowercase label

rename_lowecase_label = {
    'Positif': 'positive',
    'Netral': 'neutral',
    'Negatif': 'negative',
}

# rename lowercase label
dataset_keyword_banjir['label'] = dataset_keyword_banjir['label'].map(rename_lowecase_label)
dataset_keyword_banjir
```

# Integration Data File Tweet Keyword Jokowi

```python
dataset_keyword_jokowi
```

```python
# rename column text and label

dataset_keyword_jokowi = dataset_keyword_jokowi[['text','Label']].rename(columns={'text':'Tweet',
                                                                             'Label' : 'label'})
dataset_keyword_jokowi
```

```python
# count label

dataset_keyword_jokowi['label'].value_counts()
```

```python
# mapping label

mapping_typo_label_dataset_keyword_jokowi = {
    'positif'       : 'positive',
    'positi'        : 'positive',
    'netral'        : 'neutral',
    'netr'          : 'neutral',
    'negatif'       : 'negative',
    'negartif'      : 'negative',
    'Negatif'       : 'negative',
}

# rename lowercase label
dataset_keyword_jokowi['label'] = dataset_keyword_jokowi['label'].map(mapping_typo_label_dataset_keyword_jokowi)
dataset_keyword_jokowi
```

```python
# count label

dataset_keyword_jokowi['label'].value_counts()
```

# Integration Data File Tweet Pilpres 2024

```python
dataset_pilpres
```

```python
# drop coloumn Text_Eng

dataset_pilpres.drop(columns=["Text_Eng"], inplace=True)

# rename column Text_Ind to Tweet
dataset_pilpres.rename(columns={"Text_Ind": "Tweet"}, inplace=True)

dataset_pilpres
```

```python
# count label 

label_counts_dataset_pilpres = dataset_pilpres['label'].value_counts()
label_counts_dataset_pilpres
```

```python
# mapping label typo

mapping_typo_label_dataset_pilpres = {
    'Positive': 'positive',
    'Negative': 'negative',
}

# rename lowercase label
dataset_pilpres['label'] = dataset_pilpres['label'].map(mapping_typo_label_dataset_pilpres)
dataset_pilpres
```

```python
# count label 

label_counts_dataset_pilpres = dataset_pilpres['label'].value_counts()
label_counts_dataset_pilpres
```

# Integration Data File Tweet PPKM from Sentiment Label Number to Text

```python
dataset_ppkm
```

```python
# drop coloumn date and user

dataset_ppkm.drop(columns=["Date", "User"], inplace=True)

# rename column sentiment to label
dataset_ppkm.rename(columns={"sentiment": "label"}, inplace=True)

dataset_ppkm
```

```python
# mapping number label sentiment to text label sentiment

number_to_sentiment = {
    0: 'positive',
    1: 'neutral',
    2: 'negative'
}

# change number label sentiment to text label sentiment

dataset_ppkm['label'] = dataset_ppkm['label'].map(number_to_sentiment)
dataset_ppkm
```

# Merge DataFrame 1, 2, 3, 4, 5

```python
# merge dataframe 

dataset_training = pd.concat([dataset_tweet_emotion, dataset_keyword_banjir, dataset_keyword_jokowi, dataset_pilpres, dataset_ppkm], ignore_index=True)
dataset_training
```

```python
# count label
dataset_training['label'].value_counts()
```

# Save DataFrame dataset_training to CSV

```python
dataset_training.to_csv('dataset_training.csv', index=False)
```
