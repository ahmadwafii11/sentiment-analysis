# Read file csv data training

```python
# read file csv
import pandas as pd


dataset_emotion_tweet = pd.read_csv('Twitter_Emotion_Dataset.csv')
dataset_tweet_banjir = pd.read_csv('data_twitter.csv')
dataset_tweet_ppkm = pd.read_csv('INA_TweetsPPKM_Labeled_Pure.csv', sep=None, engine='python')
```

# Integration Data File Emotion Tweet to Sentiment Tweet

```python
dataset_emotion_tweet
```

```python
# rename column

dataset_emotion_tweet = dataset_emotion_tweet[['tweet','label']].rename(columns={'tweet':'Tweet'})
dataset_emotion_tweet
```

```python
# mapping emotion to sentiment

emotion_to_sentiment = {
    'happy': 'positive',
    'love': 'positive',
    'anger': 'negative',
    'sadness': 'negative',
    'fear': 'negative'
}

# change label emotion to sentiment
dataset_emotion_tweet['label'] = dataset_emotion_tweet['label'].map(emotion_to_sentiment)
dataset_emotion_tweet
```

# Integration Data File Tweet PPKM from Sentiment Label Number to Text

```python
dataset_tweet_ppkm
```

```python
# drop coloumn date and user

dataset_tweet_ppkm.drop(columns=["Date", "User"], inplace=True)

# rename column sentiment to label
dataset_tweet_ppkm.rename(columns={"sentiment": "label"}, inplace=True)

dataset_tweet_ppkm
```

```python
# count label 

label_counts = dataset_tweet_ppkm['label'].value_counts()
label_counts
```

```python
# Mengambil 70% dari baris dengan label 'neutral'

neutral_rows = dataset_tweet_ppkm[dataset_tweet_ppkm['label'] == 1]
reduced_neutral_rows = neutral_rows.sample(n=int(len(neutral_rows)*0.3), random_state=1)

# Mengambil semua baris dengan label selain 'neutral'

other_rows = dataset_tweet_ppkm[dataset_tweet_ppkm['label'] != 1]

# Menggabungkan baris yang telah direduksi dengan baris lainnya

dataset_tweet_ppkm = pd.concat([other_rows, reduced_neutral_rows])

dataset_tweet_ppkm
```

```python
# mapping number label sentiment to text label sentiment

number_to_sentiment = {
    0: 'positive',
    1: 'neutral',
    2: 'negative'
}

# change number label sentiment to text label sentiment

dataset_tweet_ppkm['label'] = dataset_tweet_ppkm['label'].map(number_to_sentiment)
dataset_tweet_ppkm
```

# Integration Data File Tweet Banjir Sumatra

```python
dataset_tweet_banjir
```

```python
# rename column chat and rate

dataset_tweet_banjir = dataset_tweet_banjir[['chat','rate']].rename(columns={'chat':'Tweet',
                                                                             'rate' : 'label'})
dataset_tweet_banjir
```

```python
# mapping lowercase label

rename_lowecase_label = {
    'Positif': 'positive',
    'Netral': 'neutral',
    'Negatif': 'negative',
}

# rename lowercase label
dataset_tweet_banjir['label'] = dataset_tweet_banjir['label'].map(rename_lowecase_label)
dataset_tweet_banjir
```

# Merge dataframe 1, 2, 3

```python
# merge dataframe 

dataset_training = pd.concat([dataset_emotion_tweet, dataset_tweet_ppkm, dataset_tweet_banjir], ignore_index=True)
dataset_training
```

```python
# count label
dataset_training['label'].value_counts()
```

# Save DataFrame dataset_training to CSV

```python
dataset_training.to_csv('dataset_training.csv', index=False)
```
